```python
import cv2 
import uuid
import os
import time
import csv
```

```python
IMAGES_PATH = 'data/collectedimages'
```

```python
# Define the file path and number of images
file_path = 'labels.csv'
number_imgs = 100  # You can adjust this number as needed

# Read labels from the CSV file
labels = []
with open(file_path, newline='') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)  # Skip the first row (column headings)
    for row in reader:
        labels.extend(row)

# Ensure the number of images is not greater than the number of labels
if number_imgs > len(labels):
    number_imgs = len(labels)

# Use the first `number_imgs` labels for training
training_labels = labels[:number_imgs]

# Print the training labels and the number of images
print("Training Labels:", training_labels)
print("Number of Images for Training:", len(training_labels))
```

```python
# Loop through each label
for label in labels:
    # Create a directory for the current label if it doesn't exist
    label_dir = os.path.join(IMAGES_PATH, label)
    if not os.path.exists(label_dir):
        os.makedirs(label_dir)

    # Initialize the webcam
    cap = cv2.VideoCapture(0)
    print('Collecting images for {}'.format(label))
    time.sleep(5)  # Give some time to adjust before capturing images

    # Capture 10 images for the current label
    for imgnum in range(10): 
        print('Collecting image {}'.format(imgnum))
        ret, frame = cap.read()
        
        # Generate a unique image name using UUID
        imgname = os.path.join(label_dir, '{}_{}.jpg'.format(label, str(uuid.uuid1())))
        
        # Save the captured frame as an image
        cv2.imwrite(imgname, frame)
        
        # Display the captured frame (optional)
        cv2.imshow('frame', frame)
        cv2.waitKey(2000)  # Display each image for 2 seconds

    # Release the webcam after capturing images for the current label
    cap.release()

# Close all OpenCV windows
cv2.destroyAllWindows()
```
